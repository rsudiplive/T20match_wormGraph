```
def get_inn_score(inn_del):

  inn_df = pd.DataFrame(columns=['delivery','runs_scored'])

  for delivery in inn_del:
    delivery_no = list(delivery.keys())[0]
    extras = delivery[delivery_no]['runs']['extras']
    total = delivery[delivery_no]['runs']['total']
    # print(list(delivery.keys())[0])
    # print(delivery[delivery_no])
    # print(delivery[delivery_no].keys())

    # first_inn_df = first_inn_df.append(pd.DataFrame(delivery_no,total))
    # print(int(delivery_no))
    # print(delivery_no,total)
    # total_runs = total_runs+total
    # print(total_runs)
    inn_df = inn_df.append({'delivery':delivery_no,'runs_scored': total},ignore_index=True)

  inn_df['over'] = inn_df.apply(lambda x: int(x['delivery']),axis=1)
  inn_df['over'] = inn_df['over'] +1

  inn_runs_df= inn_df.groupby('over').agg({'runs_scored':sum}).reset_index()
  inn_runs_df['Total_Score'] = inn_runs_df['runs_scored'].cumsum()

  
  
  return inn_runs_df
```

```
def get_inn_score(inn_del):

  inn_df = pd.DataFrame(columns=['delivery','runs_scored','fall_of_wickets'])

  for delivery in inn_del:
    delivery_no = list(delivery.keys())[0]
    extras = delivery[delivery_no]['runs']['extras']
    total = delivery[delivery_no]['runs']['total']
    # print(list(delivery.keys())[0])
    # print(delivery[delivery_no])
    # print(delivery[delivery_no].keys())
    wickets = 0
    if 'wicket' in list(delivery[delivery_no].keys()):
      
      wickets = wickets +1

    else:
      wickets = wickets

    # first_inn_df = first_inn_df.append(pd.DataFrame(delivery_no,total))
    # print(int(delivery_no))
    # print(delivery_no,total)
    # total_runs = total_runs+total
    # print(total_runs)
    inn_df = inn_df.append({'delivery':delivery_no,'runs_scored': total, 'fall_of_wickets': wickets},ignore_index=True)

  inn_df['over'] = inn_df.apply(lambda x: int(x['delivery']),axis=1)
  inn_df['wickets_fall'] = inn_df.apply(lambda x: int(x['fall_of_wickets']),axis=1)
  inn_df['over'] = inn_df['over'] +1

  inn_runs_df= inn_df.groupby('over').agg({'runs_scored':sum,'wickets_fall': sum}).reset_index()
  inn_runs_df['Total_Score'] = inn_runs_df['runs_scored'].cumsum()

  
  
  return inn_runs_df
```

```
first_inn_stats_df = get_inn_score(first_inn_del)
```

```
second_inn_stats_df=get_inn_score(second_inn_del)
```

```
first_inn_stats_df['innings'] = 'first'
second_inn_stats_df['innings'] = 'second'

first_inn_stats_df['team'] = first_inn_dict['1st innings']['team']
second_inn_stats_df['team'] = second_inn_dict['2nd innings']['team']
```
